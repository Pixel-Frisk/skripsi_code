```python
data = pd.read_csv('PRDECT-ID Dataset.csv', sep=',',encoding='utf-8')
data.head()
```

```python
def case_folding(comment):
    comment = comment.lower()
    return comment

data['Customer Review'] = data['Customer Review'].apply(case_folding)
data.head()
```

```python
def cleansing(comment):
    comment = comment.strip(" ")
    comment = re.sub(r'[\U0001F600-\U0001F64F]', ' ', comment) # remove emoticon
    comment = re.sub(r'[^\w\s]', ' ', comment) #remove punctuation
    comment = re.sub(r'\d', ' ', comment) #remove number
    return comment

data['Customer Review'] = data['Customer Review'].apply(cleansing)
data.head()
```

```python
def token(comments):
    tokens = word_tokenize(comments)
    return(tokens)

data['Customer Review'] = data['Customer Review'].apply(token)
data.head()
```

```python
def stopword_removal(comments):
    filtering = stopwords.words('indonesian','english')
    x = []
    data = []
    def myFunc(x):
        if x in filtering:
            return False
        else:
            return True
    fit = filter(myFunc, comments)
    for x in fit:
        data.append(x)
    return data

data['Customer Review'] = data['Customer Review'].apply(stopword_removal)
data.head()
```

```python
def stemming(comments):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in comments:
        dt = stemmer.stem(w)
        do.append(dt)
    print(do)
    d_clean = []
    d_clean = ' '.join(do)
    return d_clean
```

```python
data['Customer Review'] = data['Customer Review'].apply(stemming)
data.to_csv('data_emotion_cleanssss.csv', columns=['Customer Review', 'Emotion'], index=False)
```
